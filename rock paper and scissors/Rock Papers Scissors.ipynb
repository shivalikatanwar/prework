```python
import random
```

```python
def random_option(op):
    return random.choice(op)

def ask_option(option):
    op = input("Your choice :")
    while op not in option:
        op = input("Wrong Answer!!! Please enter your choice (btw "+str(option)+") :") 
    return op

def resolve_combat(h, m):
    combat = {"Tie": 0, "Machine Wins":1, "Human Wins":2}
    if (m == "stone" and h == "scissors") or (m == "scissors" and h == "paper") or (m=="paper" and h=="stone"):
        return combat["Machine Wins"]
    elif (h == "stone" and m == "scissors") or (h == "scissors" and m == "paper") or (h=="paper" and m=="stone"):
        return combat["Human Wins"]
    else:
        return combat["Tie"]
    
def game(player1, player2, point_player1, point_player2):
    print ("Choice of Human : ", player1)
    print ("Choice of Machine : ", player2)
    record = resolve_combat(player1, player2)
    if record == 2:
        point_player1 += record
    elif record == 1:
        point_player2 += record
    print ("State of the game => Human points:", str(point_player1), " ; Machine points:", point_player2)
    return [point_player1, point_player2]
```
